```python
# We can choose to either fill the missing values with mean/median
# depending upon skewness of the data. Or we can drop the rows with
# missing values. Or we can apply regression techniques to fill the 
# missing values, whichever applicable.
new_data = data.dropna()
```

# Handling categorical data

```python
#Convert the object data type to categorical and then assign
#the categorical or numeric values 

new_data[new_data.select_dtypes(['object']).columns] = new_data.select_dtypes(['object']).apply(lambda x: x.astype('category').cat.codes)
```
